# IMPOTAÇÃO DA BASE DE DADOS E NORMALIZAÇÃO

```python
import pandas as pd
```

```python
# Dados organizados como dicionário
dados = pd.read_json('/content/dados_hospedagem.json')
dados
```

```python
# Dados Normalizados
dados_normalizados = pd.json_normalize(dados['info_moveis'])
dados_normalizados
```

RETIRANDO AS LISTAS

```python
colunas = list(dados_normalizados.columns)
colunas
```

```python
dados_normalizados = dados_normalizados.explode(colunas[3:])
dados_normalizados
```

```python
# Resetando os Indexes para tirar a duplicacao
dados_normalizados.reset_index(drop=True, inplace=True)
dados_normalizados
```

```python
# Dados como Object, necessidade de tratamento
dados_normalizados.info()
```

# **CONVERTENDO DADOS NUMÉRICOS**



```python
import numpy as np
```

```python
# Convertendo para int
dados_normalizados['max_hospedes']= dados_normalizados['max_hospedes'].astype(np.int64)
```

```python
col_numericas = ['quantidade_banheiros','quantidade_quartos','quantidade_camas']
```

```python
dados_normalizados[col_numericas] = dados_normalizados[col_numericas].astype(np.int64)
```

```python
# Convertendo para float
dados_normalizados['avaliacao_geral'] = dados_normalizados['avaliacao_geral'].astype(np.float64)
```

```python
dados_normalizados.info()
```

```python
# Realizando o tratamento da coluna Preco
dados_normalizados['preco'] = dados_normalizados['preco'].apply(lambda x: x.replace('$','').replace(',','').strip())
```

```python
# Convertendo
dados_normalizados['preco'] = dados_normalizados['preco'].astype(np.float64)
```

**Método Applymap( ) ✔️**

Usado para converter utilizando lamba multiplas colunas

```python
# Tratando as colunas taxa_deposito e taxa_limpeza da mesma forma
dados_normalizados[['taxa_deposito','taxa_limpeza']]= dados_normalizados[['taxa_deposito','taxa_limpeza']].applymap(lambda x: x.replace('$','').replace(',','').strip())
dados_normalizados
```

```python
# Convertendo as colunas para float64
dados_normalizados[['taxa_deposito','taxa_limpeza']] = dados_normalizados[['taxa_deposito','taxa_limpeza']].astype(np.float64)
```

```python
dados_normalizados.info()
```

# **TRABALHANDO COM DADOS TEXTUAIS**

```python
# Deixando todos os textos em minúsculo
dados_normalizados['descricao_local'].str.lower()
```

```python
dados_normalizados['descricao_local'] = dados_normalizados['descricao_local'].str.lower()
```

**Método Regex ✔️**

Utilizado para fazer busca, validação e manipulação de dados




```python
dados_normalizados['descricao_local'] = dados_normalizados['descricao_local'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex=True)
dados_normalizados['descricao_local']

# Obs.: Toda essa expressão diz: "selecione todos os caracteres,
# exceto as letras (maiúsculas ou minúsculas), números, hífen e apóstrofo".
```

```python
# Retirando hifens soltos no texto
dados_normalizados['descricao_local'] = dados_normalizados['descricao_local'].str.replace('(?<!\w)-(?!\w)', '', regex=True)
dados_normalizados['descricao_local'].str.replace('(?<!\w)-(?!\w)', '', regex=True)

# Lookbehind --> (?<!\w): verifica se há um caractere de palavra ANTES do hífen
# Lookahead --> (?!\w): erifica se há um caractere de palavra DEPOIS do hífen.
```

**Método Split( ) ✔️**

Transforma textos em listas





```python
# Tratando as colunas descricao_local e comodidades
dados_normalizados['descricao_local'].str.split()
```

```python
dados_normalizados['descricao_local'] = dados_normalizados['descricao_local'].str.split()
dados_normalizados
```

```python
dados_normalizados['comodidades'].str.replace('\{|}|\"','',regex=True)
```

```python
dados_normalizados['comodidades'] = dados_normalizados['comodidades'].str.replace('\{|}|\"','',regex=True)
```

```python
dados_normalizados['comodidades'] = dados_normalizados['comodidades'].str.split(',')
dados_normalizados
```

```python
# Encontrando a quantidade de vagas por mes utilizando groupby
subset = dados_moveis.groupby(dados_moveis['data'].dt.strftime('%Y-%m'))['vaga_disponivel'].sum()
subset
```
